```python
from astropy.io import fits
from astropy.time import Time

# open file
f = fits.open('data/20190101/desi-00000003.fits.fz')

# read the time in isot
time = f[0].header['DATE-OBS']

# declare the format
t = Time(time, format='isot', scale='utc')

# Convert to MJD
MJD =  t.mjd

print(MJD)
```
